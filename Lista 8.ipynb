## Problema 1

```python
def f(Vin,Vout,RC):
    return (Vin - Vout)/RC
```

```python
#a)
RC = [0.01, 0.1, 1]
a,b = 0, 10
N = 1000
h = (b-a)/N
t = np.arange(a,b,h)
```

```python
for q in range(0,3):
    temp = 0
    Vout = []
    for i in range(0,N):
        Vout.append(temp)
        
        if((2*t[i])>int(2*t[i])):
            if(int(2*t[i])%2==0):
                Vin = 1
            else:
                Vin = -1
                
        else:
            if((int(2*t[i])-1)%2==0):
                Vin = 1
            else:
                Vin = -1
                
        K1 = h*f(Vin,temp, RC[q])
        K2 = h*f(Vin + h/2, temp + K1/2, RC[q])
        K3 = h*f(Vin + h/2, temp + K2/2, RC[q]) 
        K4 = h*f(Vin + h, temp + K3, RC[q])
        
        temp = temp + (K1 + 2*K2 + 2*K3 + K4)/6
        
        
    plt.figure(figsize = (8,5))
    plt.plot(t, Vout)
    plt.xlabel("t")
    plt.ylabel("Vout(t)")
```

## Problema 2

```python
alfha = 1
beta = 0.5
lamb = 0.5
delta = 2

def g(t, r):
    x, y = r[0], r[1]
    X = alfha*x - beta*x*y
    Y = lamb*x*y - delta*y
    return np.array([X, Y])

def rk4(t, r):
    K1 = h*g(t,r)
    K2 = h*g(t + h/2, r + K1/2)
    K3 = h*g(t + h/2, r + K2/2) 
    K4 = h*g(t + h, r + K3)
    return r + (K1 + 2*K2 + 2*K3 + K4)/6
```

```python
#a)
a, b = 0, 30
N =3000
h = (b-a)/N
t = np.arange(a,b,h)
x, y = [], []
r = np.array([2,2], float)

for i in range(0,N):
    
    x.append(r[0])
    y.append(r[1])
    r = rk4(t[i],r)
    
plt.figure(figsize = (8,5))
plt.plot(t,x,'r', label= "Coelhos")
plt.plot(t,y,'b', label = "Raposas")
plt.xlabel("")
plt.legend()
```

### b
Da para perceber que quando a população de coelhos chega no mínimo, a população de raposas está caindo, e isso faz com que a população de coelhos volte a crescer de novo. E é perceptível que quando a população de coelhos volta a crescer depois de um certo tempo a população de raposas volta a crescer também, e com a população de raposas crescendo a de coelhos volta a cair de novo, e assim se ambos se comportando como um ciclo. Sendo assim, é possível perceber que esses gráficos de população de coelhos e de raposas se comportam como um Movimento Harmônico simples.
