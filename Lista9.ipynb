## Problema 1

```python
#a)
l = 0.1
C = 2 
omega = 5
g = 9.81
#z=dteta/dt

def f(t,  r):
    teta, z = r[0], r[1]
    fteta = z
    fz = sin(teta)*(-g/l) + C*cos(teta)*sin(omega*t)
    
    return array([fteta, fz], float)

def rk4(t, r):
    K1 = h*f(t,r)
    K2 = h*f(t + h/2, r + K1/2)
    K3 = h*f(t + h/2, r + K2/2) 
    K4 = h*f(t + h, r + K3)
    return r + (K1 + 2*K2 + 2*K3 + K4)/6
```

```python
a, b = 0, 100
N = 100000
h = (b-a)/N
t = np.arange(a, b, h)
teta, z = [], []
r = array([0,0], float)

for i in t:
    teta.append(r[0])
    z.append(r[1])
    r = rk4(i, r)
    
figure(figsize=(8,5))
plot(t,teta)
xlabel("t")
ylabel("teta(t)")
```

```python
#Extra
figure(figsize=(8,5))
plot(teta,z)
xlabel("teta(t)")
ylabel("dteta/dt")
title("Grfico do espao de fase do pndulo forado")
```

```python
#b)
m = 1
R = 0.08
p = 1.22
C = 0.47
g = 9.81
def f(t,  r):
    x, y, vx, vy = r[0], r[1], r[2], r[3]
    
    fx = vx
    fy = vy
    m2 = 2*m
    fvx = ((-(pi*(R**2)*p*C))/m2)*vx*sqrt(vx**2 + vy**2)
    fvy = -g-(((pi*(R**2)*p*C)/m2)*vy*sqrt(vx**2 + vy**2))
    return array([fx, fy, fvx, fvy], float)

def rk4(t, r):
    K1 = h*f(t,r)
    K2 = h*f(t + h/2, r + K1/2)
    K3 = h*f(t + h/2, r + K2/2) 
    K4 = h*f(t + h, r + K3)
    return r + (K1 + 2*K2 + 2*K3 + K4)/6
```

```python
#b)
G = 1
M = 10
L = 2
def f(t, r):
    x, y, vx, vy = r[0], r[1], r[2], r[3]
    fx = vx
    fy = vy
    r = sqrt(x**2 + y**2)
    fvx = -G*M*x/((r**2)*sqrt(r**2 + (L**2)/4))
    fvy = -G*M*y/((r**2)*sqrt(r**2 + (L**2)/4))
    return array([fx, fy, fvx, fvy], float)

def rk4(t, r):
    K1 = h*f(t,r)
    K2 = h*f(t + h/2, r + K1/2)
    K3 = h*f(t + h/2, r + K2/2) 
    K4 = h*f(t + h, r + K3)
    return r + (K1 + 2*K2 + 2*K3 + K4)/6
```

```python
a, b = 0, 10
N = 100000
h = (b-a)/N
t = arange(a, b, h)
x, y, vx, vy = [], [], [], []
r = array([1, 0, 0, 1], float)
for i in t:
        
    x.append(r[0])
    y.append(r[1])
    vx.append(r[2])
    vy.append(r[3])
    r = rk4(i, r)
        
figure(figsize=(8,5))
plot(x, y)
xlabel("x(t)")
ylabel("y(t)")
title("Grfico da trajetria do Lixo espacial")
```

```python
m = 1        
R = 0.08
theta = pi/6
ro = 1.22
C = 0.47
g = 9.81
vo = 100
N = 1000
to = 0
tf = 6.6
h = (tf-to)/N

cons = (pi*R**2*ro*C)/2*m

def f(r,t):
    x = r[0]
    vx = r[1]
    y = r[2]
    vy = r[3]
    
    return array([vx, -cons*vx*sqrt(vx**2 + vy**2), vy, -g-cons*vy*sqrt(vx**2+vy**2)], float)

t_ptos = arange(to, tf, h)
eixox = []
eixoy = []

r = array([0, vo*cos(theta), 0, vo*sin(theta)], float)

#Mtodo Runge-Kutta p/ 4 ordem:
for t in t_ptos:
    eixox.append(r[0])
    eixoy.append(r[2])
    
    k1 = h * f(r, t)
    k2 = h * f(r + 0.5 * k1, t + 0.5 * h)
    k3 = h * f(r + 0.5 * k2, t + 0.5 * h)
    k4 = h * f(r + k3, t + h)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    
plot(eixox,eixoy)
```
