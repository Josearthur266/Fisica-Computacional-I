## Problema 1

```python
def f(x):
    return 1 + tanh(2*x)/2
def der(x):
    h = 1e-5
    return (f(x+h/2)-f(x-h/2))/h
def exat(x):
    return 1/((cosh(2*x))**2)
```

```python
x = np.linspace(-2,2,50)

deri = []
for i in range(0,50):
    deri.append(der(x[i]))
exato = []
for i in range(0,50):
    exato.append(exat(x[i]))
    
fig =  plt.figure(figsize=(5,3))
plt.plot(x,deri, '-g')
plt.plot(x,exato,'.b')
```

## Problema 3

```python
def f(x):
    return 1 + tanh(2*x)/2
def second_dev(x):
    return (f(x+h)-2*f(x)+f(x-h))/(h**2)
def exato(x):
    return -4*(tanh(2*x)/(cosh(2*x))**2)
```

```python
x = np.linspace(-2,2,100)
der = []
ext = []
for i in range(0,100):
    der.append(second_dev(x[i]))
for i in range(0,100):
    ext.append(exato(x[i]))
    
fig = plt.figure(figsize=(5,3))    
plt.plot(x, der, '.b')
plt.plot(x, ext, '-g')
```

## Problema 4

```python
#a)
def P(r,q):
    k = 8.98e9
    return q*k/r
q1, q2 = 1, -1
# considerei as coordenadas de q1(-5,0) e q2(5,0), lembrando que esses pontos estão em centímetros
x =  np.linspace(-0.5, 0.5, 100)
y =  np.linspace(-0.5, 0.5, 100)

Pr = []
for i in range(0,100):
    linha = []
    for j in range(0, 100):
        temp1 = ((-0.05-x[j])**2 + (y[i])**2)**0.5
        temp2 = ((0.05-x[j])**2 + (y[i])**2)**0.5
        p1 = (P(temp1,q1))
        p2 = (P(temp2,q2))
        linha.append(p1+p2)
    Pr.append(linha)
```

```python
plt.imshow(Pr, origin='lower', extent = [0,1,0,1],vmax=1e10, vmin = -1e10)
plt.colorbar()
plt.hot()
```

```python
#b)
h = 1e-5
def f(x,y,q):
    k = 8.98e9
    return q*k/((x**2 + y**2)**0.5)
def dx(x,y,q):
    return (f(x+h/2,y,q)-f(x-h/2,y,q))/h
def dy(x,y,q):
    return (f(x, y+h/2,q)-f(x,y-h/2,q))/h
```

```python
dex = []
dey= []
E = []
Ex = []
Ey = []
for i in range(0,100):
    linha = []
    linhax = []
    linhay = []
    temp = []
    temp2 = []
    temp3 = []
    temp4 = []
    for j in range(0,100):
        temp.append(dx(-0.05-x[j],y[i],q1))
        temp2.append(dy(-0.05-x[j],y[i],q1))
        temp3.append(dx(0.05-x[j],y[i],q2))
        temp4.append(dy(0.05-x[j],y[i],q2))
        linhax.append(dx(-0.05-x[j],y[i],q1)+dx(0.05-x[j],y[i],q2))
        linhay.append(dy(-0.05-x[j],y[i],q1)+dy(0.05-x[j],y[i],q2))
        linha.append(((dx(-0.05-x[j],y[i],q1)+dx(0.05-x[j],y[i],q2))**2 + (dy(-0.05-x[j],y[i],q1)+dy(0.05-x[j],y[i],q2))**2)**0.5)
    E.append(linha)
    Ex.append(linhax)
    Ey.append(linhay)
    dex.append(temp + temp3)
    dey.append(temp2 + temp4)
```

```python
plt.figure(figsize=(8,5))
plt.imshow(E, origin='lower', extent = [0,1,0,1], vmax=1e11, vmin = -1e11)
plt.colorbar()
plt.hot()
```
